```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.patches import Circle

%matplotlib inline
```

```python
fig = plt.figure(figsize = (6,6))
ax = fig.add_axes([1,1,1,1])
ax.plot([40,50,60],
        [0,20,0],'k',linewidth=8)
ax.plot([50,50,50],
        [20,55,20],'k',linewidth=8)
ax.plot([30,50,70],
        [30,40,30],'k',linewidth=8)
circle1 = Circle((80, 80), 5, edgecolor='k',facecolor='y',linestyle='--', linewidth=5)
circle2 = Circle((50, 60), 5, facecolor='skyblue')
circle3 = Circle((48, 60), 1, facecolor='k')
circle4 = Circle((52, 60), 1, facecolor='k')
ax.add_patch(circle1)
ax.add_patch(circle2)
ax.add_patch(circle3)
ax.add_patch(circle4)
ax.grid(which='major', linestyle='-', linewidth='0.5', color='black')
plt.show()
```
